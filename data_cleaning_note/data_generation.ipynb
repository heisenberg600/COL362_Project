```python
url = "https://randomuser.me/api/?results=100"
person_data=[]
for i in tqdm(range(1000)):
    payload={}
    headers = {}
    response = requests.request("GET", url, headers=headers, data=payload)
    
    person_data.extend(response.json()['results'][:])
    time.sleep(1)
```

```python
df_f=pd.read_csv('persons.csv')
df_f.set_index('id',inplace=True)
restaurats=pd.read_csv('restaurants.csv')
restaurats.set_index('restId',inplace=True)
restaurant_index=restaurats.index
person_index=df_f.index
cities=pd.read_csv('cities.csv')
review_data=[]
reviews_prob=[]
cityId=[]
for i in range(100):
    a=[]
    for i in range(5):
        a.append(random.randint(1,5))
    b=sum(a)
    a=list(map(lambda x: x/b,a))
    reviews_prob.append(a)
```

```python
for i in person_index:
    livesin=random.randint(1,len(cities)+1)
    cityId.append(livesin)
    if (df_f.at[i,'priviledge']==2):
        restaurant_city=restaurant_index[restaurats['cityId']==livesin].to_list()
        restaurant_ncity=restaurant_index[restaurats['cityId']!=livesin].to_list()
        restaurant_city_s=random.sample(restaurant_city,random.randint(0,int(len(restaurant_city)/2)))
        prob=reviews_prob[random.randint(0,len(reviews_prob)-1)]
        for j in restaurant_city_s:
            review_data.append([i,j,np.random.choice(np.arange(1,6),p=prob)])
        restaurant_ncity_s=random.sample(restaurant_ncity,random.randint(1,100))
        for j in restaurant_ncity_s:
            review_data.append([i,j,np.random.choice(np.arange(1,6),p=prob)])
review_data
```

```python
reviews=pd.DataFrame(review_data,columns=['userId','restId','stars'])
```

```python
reviews.to_csv('reviews.csv',index=False)
```

```python
avgreview=[0 for i in range(len(restaurats))]
numreview=[0 for i in range(len(restaurats))]

for index, rows in reviews.iterrows():
    avgreview[rows['restId']-1]=(avgreview[rows['restId']-1]*numreview[rows['restId']-1]+rows['stars'])/(numreview[rows['restId']-1]+1)
    numreview[rows['restId']-1]+=1

restaurats['avgReview']=list(map(lambda x: round(x,2),avgreview))
restaurats['num_reviews']=numreview
restaurats
```
